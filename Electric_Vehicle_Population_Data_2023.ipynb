# **Attribute Information**

**VIN (1-10):**
The 1st 10 characters of each vehicle's Vehicle Identification Number (VIN).


**County:**
The county in which the registered owner resides.

**City:**
The city in which the registered owner resides.

**State:**
The state in which the registered owner resides.

**Postal Code:**
The 5 digit zip code in which the registered owner resides.

**Model Year:**
The model year of the vehicle, determined by decoding the Vehicle Identification Number (VIN).

**Make:**
The manufacturer of the vehicle, determined by decoding the Vehicle Identification Number (VIN).

**Model:**
The model of the vehicle, determined by decoding the Vehicle Identification Number (VIN).

**Electric Vehicle Type:**
This distinguishes the vehicle as all electric or a plug-in hybrid.

**Clean Alternative Fuel Vehicle (CAFV) Eligibility:**
This categorizes vehicle as Clean Alternative Fuel Vehicles (CAFVs) based on the fuel requirement and electric-only range

# **4.Visualization**

**1.Which car is more recognized by the customers**

```python
#finding most popular car
max_car=df.Model.value_counts().nlargest(1)
print("Most selling car name and count of car is-",max_car)
```

```python
#top 10 cars
max_cars=df.Model.value_counts().nlargest(10)
max_cars=pd.DataFrame(max_cars).reset_index()
max_cars
```

```python
#barplot
ax=sns.barplot(data=max_cars,x="index",y="Model")

#to show lables
for bars in ax.containers:
    ax.bar_label(bars)

#lables,title
plt.xticks(rotation="vertical")
plt.xlabel("Name of cars")
plt.ylabel("Count of cars")
plt.title("Name of cars Vs Count of cars")
plt.show()
```

<font color='red'>**Conclusion: "MODEL 3" which is of "TESLA" is the more sold car**

**2.Which car company is popular in market according to selling**
